## What is a Function?

A **function** is a reusable block of code that performs a specific task. Functions help you break down complex problems into smaller, manageable pieces. They also promote code reusability, making your programs more organized and easier to maintain.

### Why Use Functions?

- **Reusability:** Write once, use multiple times.
- **Organization:** Break down complex tasks into simpler parts.
- **Maintainability:** Easier to debug and update code.
- **Readability:** Clear structure makes code easier to understand.

## Defining Your Own Functions

You can create your own functions in Python. In fact, this is one of the most powerful features of programming — the ability to create your own functions. There are three major advantages to functions that we will discuss:

* Code reuse so that we don’t have to write the same statements repeatedly.
* Easier to debug a chunk of code in isolation from the rest of the program.
* Clean decomposition to break a program down into smaller pieces without keeping the whole thing in our mind at once.

### Definition Syntax

To create a new function, you write:

1. The def keyword (which stands for "define")
2. The name of the function
3. An open parenthesis (
4. Each of the parameters, separated by commas (,)
5. A closed parenthesis )
6. A dash and a greater than (->, representing an arrow)
7. The return type of the function
8. A colon (:)


For each parameter, you write:

1. The name of the parameter
2. A colon
3. The type of the parameter

This entire line — including the parameters — is called the *function header*.

# Defining Your First Function

Let's start by creating a simple function that greets the user.

## Example: Greeting Function

```python
# Define the function
def greet():
    print("Hello, World!")

# Call the function
greet()
```

### Explanation
* `def` Keyword: Used to define a function.
* Function Name: greet is the name of the function.
* Parentheses `()`: Used to call the function.
* Colon :`:` Indicates the start of the function body.
* Indentation: Python uses indentation to define the scope of the function.

# Functions with Parameters
Functions can accept inputs, known as parameters, which allow them to perform tasks based on those inputs.

When you call a function, you are executing the code stored in the body of the function. Everything "inside" the body should be indented 4 spaces. The body must be there — in other words, it cannot be empty.

After you've defined a function, you can use it by calling the function. As we did before, we combine the name of the function with calling parentheses.

## Example: Adding Two Numbers

```python
# Define the function with parameters
def add(a, b):
    print(f"The sum of {a} and {b} is {a + b}")

# Call the function with arguments
add(3, 5)
```

```python
def add(a: int, b: int) -> int:
    print(f"The sum of {a} and {b} is {a + b}")

# Call the function with arguments
add(10, 20)
# The return statement is not needed in this case.
# This function prints the sum instead of returning it.
```

### Explanation
* Parameters: `a` and `b` are parameters that the function add accepts.
* Arguments: `3` and `5` are the arguments passed to the function when called.

# Type Hints for Parameters and Return Values

Python allows you to annotate the types of parameters and return values for better code clarity and type checking.

### Parameters

When you define a function, you can choose to add parameters to the header. These parameters will take on the value of the arguments when the function is called. Each argument exactly matches one parameter. In the code below, the parameter first will match to 3, -2, and -10. The parameter second will match to 8, 5, and 10. Remember, each function call happens one after the other.

```python
def subtract(first: int, second: int) -> int:
    return first - second

print(subtract(3, 8))  # Output: -5
print(subtract(-2, 5))  # Output: -7

result = subtract(-10, 10)
print(result)  # Output: -20
```

# Global vs. Local Scope
Understanding variable scope is crucial when working with functions. Variables can be global (accessible everywhere) or local (accessible only within the function).


## Scope
In a program, the scope of a variable indicates how long that variable is available. This is also known as the "lifetime" or "visibility" of a variable. When a program is no longer in scope, that variable is no longer available.

* "Lifetime"
* "Visibility"
* "Availability"
* "How long the variable is available"

## Global Scope
Variables defined at the top level are known as global variables. Once a variable is defined, it is available on subsequent lines. That variable lives until the end of the program.

In the program shown here, the `grade` variable starts its life on the second line and is available until the end of the program.

## Local Scope

Each function has its own local scope. Variables that are defined as parameters or within a function live until the function ends. These are local variables.

Variables defined in one function are not available outside the function. This simplifies the reading of any function - you only need to worry about things defined in the function itself.



## Example: Global and Local Variables

```python
# Global variable
message = "Hello from the global scope!"

def display_message():
    # Local variable
    message = "Hello from inside the function!"
    print(message)

# Call the function
display_message()

# Print the global variable
print(message)
```

### Explanation
* Global Variable: Defined outside any function and accessible anywhere in the code.
* Local Variable: Defined inside a function and only accessible within that function.
* Variable Shadowing: The local message variable shadows the global message within the function.

# Modifying Global Variables
To modify a global variable within a function, use the `global` keyword.

```python
count = 10

def increment():
    global count
    count += 1
    print(f"Inside function: {count}")

# Call the function
increment()

# Print the global variable
print(f"Outside function: {count}")
```

### Explanation
* `global` Keyword: Allows the function to modify the global variable count

# Testing Functions with `assert`

Testing your functions is an essential part of programming. It ensures that your functions behave as expected. One simple way to test functions in Python is by using the `assert` statement.

## What is `assert`?

The `assert` statement is used to verify that a condition is `True`. If the condition is `False`, it raises an `AssertionError`. This is useful for writing simple tests to check if your functions return the correct results.

## Why Use `assert`?
- **Simplicity**: Easy to write and understand.
- **Immediate Feedback**: Quickly identifies when a function doesn't behave as expected.
- **Documentation**: Serves as documentation for what your function is supposed to do.

## How to Use `assert`

The basic syntax of `assert` is:

```python
assert condition, "Error message if condition is False"
```


## Example: Testing the `add` Function

Let's revisit the `add` function and write some tests using `assert`.

```python
def add(a, b):
    return a + b

# Test cases
assert add(2, 3) == 5, "add(2, 3) should return 5"
assert add(-1, 1) == 0, "add(-1, 1) should return 0"
assert add(0, 0) == 0, "add(0, 0) should return 0"
assert add(100, 200) == 300, "add(100, 200) should return 300"

print("All tests passed for add function.")
```

### Explanation:
- **Test Cases**: We call `add` with different arguments and use `assert` to check if the returned value matches the expected result.
- **Success**: If all `assert` statements pass, the final `print` statement is executed.
- **Failure**: If any `assert` fails, an `AssertionError` is raised with the specified message.

## Best Practices for Using `assert`
- **Test Different Scenarios**: Include positive, negative, and edge cases.
- **Keep Tests Simple**: Each `assert` should test one condition.
- **Provide Clear Error Messages**: Helps in quickly identifying what went wrong.
- **Automate Testing**: Consider using testing frameworks like `unittest` or `pytest` for more comprehensive testing.
