## How to define Sheldon's Spot

*Sheldon sitting in his spot on the couch.* The Big Bang Theory

Sheldon defines his seat as "...the singular location in space around which revolves my entire universe."

Sheldon relates his spot to himself in mathematical terms in "The Cushion Saturation":

> "In an ever-changing world, it is a single point of consistency. If my life were expressed as a function on a four-dimensional Cartesian coordinate system, that spot, at the moment I first sat on it, would be (0,0,0,0)."


In the pilot, Sheldon tells Penny,

> "In the winter that seat is close enough to the radiator to remain warm, and yet not so close as to cause perspiration. In the summer its directly in the path of a cross breeze created by open windows there (points at a window), and there (points at another window). It faces the television at an angle that is neither direct, thus discouraging conversation, nor so far wide to create a parallax distortion, I could go on, but I think Ive made my point."
